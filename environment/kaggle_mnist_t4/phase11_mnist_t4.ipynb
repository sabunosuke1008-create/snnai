# SNNAI Phase11: MNIST SNN on T4 GPU

This notebook demonstrates Phase8 MNIST SNN training on a T4 GPU.

```python
!pip install snntorch==0.9.4 torchvision --quiet
!rm -rf snnai && git clone https://github.com/sabunosuke1008-create/snnai.git
%cd snnai
from snnai.benchmarks.data_loader import get_mnist_datasets
from snnai.benchmarks.mnist_snn import MNISTSNN, evaluate_mnist_snn, train_mnist_snn
from snnai.utils import auto_select_device
from torch.utils.data import DataLoader

device = auto_select_device()
print('device:', device)
train, test = get_mnist_datasets(root='./data')
train_loader = DataLoader(train, batch_size=128, shuffle=True)
test_loader = DataLoader(test, batch_size=128)
model = MNISTSNN(time_steps=20).to(device)
train_mnist_snn(model, train_loader, device=device, epochs=1, time_steps=20)
acc = evaluate_mnist_snn(model, test_loader, device=device, time_steps=20)
print('test accuracy:', acc)
```
